```
from yargy import Parser, rule, or_
from yargy.predicates import in_, in_caseless
from yargy.tokenizer import MorphTokenizer
from yargy.pipelines import morph_pipeline, caseless_pipeline
from yargy.interpretation import fact
from IPython.display import display
```

```
game = fact(
    'game',
    ['name', 'version', 'part']
)
```

Правило для Darksouls

```
DarkSouls = rule(
    # на атрибут name
    morph_pipeline(['Dark Souls', 'dark souls', 'DarkSouls', 'dark Souls',\
                    'DARKSOULS', 'DARK SOULS',
                    'Dark souls', 'darksouls', 'дарк соулз', 'дарк соулс',\
                    'дарксоулз', 'дарксоулс', 'Дарксоулс', 'Дарксоулз',\
                    'ДаркСоулз', 'ДаркСоулс', 'Дарк Соулз', 'Дарк Соулс',\
                    'ДАРКСОУЛЗ', 'ДАРКСОУЛС', 'ДАРК СОУЛЗ', 'ДАРК СОУЛС']).interpretation(game.name), 
    # на атрибут version
    morph_pipeline(['1', '2', '3',\
                    'I', 'II', 'III',\
                    'one', 'One', 'ONE',\
                    'two', 'Two', 'TWO',\
                    'three' 'Three', 'THREE']).interpretation(game.version) 
)
```

Правило для Assassin's Creed

```
AssassinSCreed = rule(
    morph_pipeline(['Assassin\'s Creed', 'Assassin\'sCreed', 'assassin\'s creed', 'assassin\'screed',\
                    'Assassins Creed', 'AssassinsCreed', 'assassins creed', 'assassinscreed',\
                    # '', '', '', '',\
                    'асасинс крид', 'Асасинс Крид', 'Асасинс крид', 'асасинс Крид',\
                    'асасин крид', 'Асасин Крид', 'Асасин крид', 'асасин Крид',\
                    'АСАСИНС КРИД', 'АСАСИН КРИД'\
                    'асасинскрид', 'АсасинсКрид', 'Асасинскрид', 'асасинсКрид',\
                    'асасинкрид', 'АсасинКрид', 'Асасинкрид', 'асасинКрид',\
                    'АСАСИНСКРИД', 'АСАСИНКРИД'\
                    ]).interpretation(game.name),
    morph_pipeline(['1', '2', '3', '4',\
                    'I', 'II', 'III', 'IV',\
                    'one', 'One', 'ONE',\
                    'two', 'Two', 'TWO',\
                    'three' 'Three', 'THREE',\
                    'four' 'Four', 'FOUR'\
                    ]).interpretation(game.version).optional(),
    morph_pipeline(['синдикат', 'Синдикат', 'СИНДИКАТ',\
                   'syndicate', 'Syndicate', 'SYNDICATE',\
                   'Black Flag', 'black flag', 'BLACK FLAG',\
                   'Black flag', 'black Flag', 'блэк флаг',\
                   'Блэк Флаг', 'Блэк флаг', 'блэк Флаг', 'БЛЭК ФЛАГ',\
                   'BlackFlag', 'blackflag', 'BLACKFLAG',\
                   'Blackflag', 'blackFlag', 'блэкфлаг',\
                   'БлэкФлаг', 'Блэкфлаг', 'блэкФлаг', 'БЛЭКФЛАГ',\
                   'Unity', 'unity', 'UNITY', 'Единство',\
                   'единство', 'ЕДИНСТВО',\
                   'Единение', 'единение', 'ЕДИНЕНИЕ'\
                   ]).interpretation(game.part).optional()
)
```

Правило для GTA

```
GTA = rule(
    morph_pipeline(['GTA', 'gta', 'Gta', 'ГТА', 'гта', 'Гта', 'Grand Theft Auto',\
                    'GrandTheftAuto', 'grandtheftauto', 'grand theft auto', 'GRAND THEFT AUTO',\
                    'GRANDTHEFTAUTO', 'Grandtheftauto', 'Grand theft auto',\
                    # '', '', '', '',\
                    ]).interpretation(game.name),
    morph_pipeline(['1', '2', '3', '4', '5',\
                    'I', 'II', 'III', 'IV', 'V',\
                    'one', 'One', 'ONE',\
                    'two', 'Two', 'TWO',\
                    'three' 'Three', 'THREE',\
                    'four' 'Four', 'FOUR',\
                    'five' 'Five', 'FIVE'\
                    ]).interpretation(game.version)    
    )
```

Общее правило

```
GAME = or_(DarkSouls, GTA, AssassinSCreed).interpretation(game)
parser = Parser(GAME) 
```

```
matches = []

for sent in data.text: #[:100]
    for match in parser.findall(sent):
        matches.append(match.fact)
```

```
print(len(matches))
```

```
for m in matches:
    print(m.name, m.version)
```
