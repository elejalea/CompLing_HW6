regex позволяет писать выражения с nested set `[[a-z]--[kw]]`

```python
import regex
s1 = 'kakahuete, cacahuete, wagon, vagon'
res = regex.findall(r'\b([[a-z]--[kw]]+)\b', s1, flags=regex.VERSION1)
print(res)
```
